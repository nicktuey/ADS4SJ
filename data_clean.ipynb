```python
#Group all the more insignicicant incident descriptions into an 'Other' Category (Anything that doesn't represent more than 1% of the data)
def collapse_rare_incidents(df, col_name='nature', threshold=0.01):
    # Calculate value counts normalized to get proportions
    value_counts = df[col_name].value_counts(normalize=True)
    
    # Find incident types that occur more than the threshold (e.g., 1%)
    valid_types = value_counts[value_counts > threshold].index
    
    # Apply the change
    df[col_name] = df[col_name].apply(lambda x: x if x in valid_types else 'OTHER')
    
    return df
```

```python
features = collapse_rare_incidents(features)
```

```python
features['nature'] = features['nature'].str.replace('SUICIDAL SUBJECT, CAHOOTS', 'SUICIDAL SUBJECT')
features['nature'] = features['nature'].str.replace('TRANSPORT, CAHOOTS', 'TRANSPORT')
features['nature'] = features['nature'].str.replace('ASSIST PUBLIC- POLICE', 'ASSIST PUBLIC')
features['nature'] = features['nature'].str.replace('PUBLIC ASSIST, CAHOOTS', 'ASSIST PUBLIC')
features['nature'] = features['nature'].str.replace('CHECK WELFARE, CAHOOTS', 'CHECK WELFARE')
features['nature'] = features['nature'].str.replace('ASSIST PD, CAHOOTS', 'ASSIST PD')
```
